<h2 style = "font-size:3rem;color:orange">Amplitude Modulation</h2>

AM: altering the amplitude of an oscillator. In the simplest case there is one $carrier$ and one $modulator$:

$y = \cos(f_c) \cdot \{A_m \cdot \cos(f_m) + A_c\}$


```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd 

import librosa                   
import librosa.display   
```

Setup carrier and modulator wave and perform AM

```python
sr = 44100 # sampling rate
dur = 1    # sample duration in sec

num_samples = sr * dur
t  = np.linspace(0,dur,num_samples)

fc = 1000
Ac = 0.5
fm = 200
Am = 0.8

y = (Ac + Am * np.cos(2*np.pi*fm*t)) * np.cos(2*np.pi*fc*t)
```

Plot results in time and frequency domain

```python
freq_res = 1
N = int(sr / freq_res) # nuumber of DFT points
T = sr / N

tY = np.linspace(0, (N -1) * T, num=N)  # time steps
Y = 2 * np.fft.fft(y,N) / N             # normalized magnitude 
Y[0] /= 2                               # DC value not doubled


plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.title('Time representation')
plt.xlabel('Time in sec')
plt.ylabel('Amplitude')
plt.plot(t[0:500], y[0:500])

plt.subplot(1,2,2)
plt.title('Frequency representation')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
# plot first 2000 Hz: 
plt.plot(tY[0:int(2000 * sr / (N))], abs(Y[0:int(2000 * sr / (N))]),  color='indigo')
```

Listen

```python
ipd.Audio(data = y, rate = sr)
```

<h2 style = "font-size:3rem;color:orange">Ring Modulation</h2>

RM is similar to AM, but the carrier's amplitude component is excluded

$y = \cos(f_c) \cdot A_m \cdot \cos(f_m)$

```python
y = Am * np.cos(2*np.pi*fm*t) * np.cos(2*np.pi*fc*t)
```

Plotting the results again. Note the missing carrier frequency

```python
freq_res = 1
N = int(sr / freq_res) # nuumber of DFT points
T = sr / N

tY = np.linspace(0, (N -1) * T, num=N)  # time steps
Y = 2 * np.fft.fft(y,N) / N             # normalized magnitude 
Y[0] /= 2                               # DC value not doubled


plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.title('Time representation')
plt.xlabel('Time in sec')
plt.ylabel('Amplitude')
plt.plot(t[0:500], y[0:500])

plt.subplot(1,2,2)
plt.title('Frequency representation')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
# plot first 2000 Hz: 
plt.plot(tY[0:int(2000 * sr / (N))], abs(Y[0:int(2000 * sr / (N))]),  color='indigo')
```

```python
ipd.Audio(data = y, rate = sr)
```

<h2 style = "font-size:3rem;color:orange">Frequency Modulation</h2>

In the case of FM, the frequency of the carrier is modulated:

$A_c \cdot \cos(f_c + A_m \cdot \cos(f_m))$

```python
y = Ac * np.cos(2*np.pi*fc*t + Am * np.cos(2*np.pi*fm*t))
```

```python
freq_res = 1
N = int(sr / freq_res) # nuumber of DFT points
T = sr / N

tY = np.linspace(0, (N -1) * T, num=N)  # time steps
Y = 2 * np.fft.fft(y,N) / N             # normalized magnitude 
Y[0] /= 2                               # DC value not doubled


plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.title('Time representation')
plt.xlabel('Time in sec')
plt.ylabel('Amplitude')
plt.plot(t[0:500], y[0:500])

plt.subplot(1,2,2)
plt.title('Frequency representation')
plt.xlabel('Frequency in Hz')
plt.ylabel('Magnitude')
# plot first 2000 Hz: 
plt.plot(tY[0:int(2000 * sr / (N))], abs(Y[0:int(2000 * sr / (N))]),  color='indigo')
```

```python
ipd.Audio(data = y, rate = sr)
```
